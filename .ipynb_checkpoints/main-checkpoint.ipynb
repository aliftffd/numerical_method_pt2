```python
import numpy as np
from scipy.optimize import fsolve, brentq
import matplotlib.pyplot as plt
```

```python

# Define the parameters (you'll need actual values)
wavelength = 0.1  # Example wavelength (meters)
antenna_height = 10  # Example antenna height (meters)
target_height = 5  # Example target height (meters)
range_val = 1000  # Example range (meters)
n = 0  # First destructive interference

# Define the function whose root we want to find
def destructive_interference_eq(theta):
    phase_diff = (4 * np.pi * antenna_height * target_height / (wavelength * range_val)) * np.sin(theta)
    return phase_diff - (2 * n + 1) * np.pi

# Initial guess for fsolve (in radians)
initial_guess = 0.01

# Solve using fsolve
destructive_angle_fsolve = fsolve(destructive_interference_eq, initial_guess)
print(f"Destructive interference angle (fsolve, n={n}): {np.degrees(destructive_angle_fsolve[0]):.2f} degrees")

```

```python
# Solve using brentq (requires finding an interval where the function changes sign)
# You might need to plot the function to find such an interval
theta_range = np.linspace(0, np.pi / 2, 100)
phase_diff_values = destructive_interference_eq(theta_range)

# Find where the function crosses zero (approximately)
sign_change_indices = np.where(np.diff(np.sign(phase_diff_values)))[0]
if len(sign_change_indices) > 0:
    lower_bound = theta_range[sign_change_indices[0]]
    upper_bound = theta_range[sign_change_indices[0] + 1]
    try:
        destructive_angle_brentq = brentq(destructive_interference_eq, lower_bound, upper_bound)
        print(f"Destructive interference angle (brentq, n={n}): {np.degrees(destructive_angle_brentq):.2f} degrees")
    except ValueError:
        print("Could not find a suitable interval for brentq.")

# Explanation of Iteration Steps:
print("\nExplanation of Iteration Steps:")
print("Numerical root-finding algorithms like fsolve and brentq use iterative methods to approximate the root of the equation.")
print("  - fsolve: Starts with an initial guess and uses algorithms (e.g., Newton-Raphson or its variations) to iteratively refine the guess until the function value is very close to zero.")
print("  - brentq: Requires an initial interval where the function has opposite signs at the endpoints. It then iteratively narrows down this interval while guaranteeing convergence to a root within that interval. It combines the robustness of the bisection method with the faster convergence of secant-like methods.")
print("The number of iterations depends on the algorithm's convergence rate and the desired tolerance.")

# Optional: Plot the function to visualize the root
plt.plot(np.degrees(theta_range), phase_diff_values)
plt.axhline(0, color='black', linestyle='--')
plt.xlabel("Angle (degrees)")
plt.ylabel("Phase Difference - (2n+1)π")
plt.title("Root Equation for Destructive Interference")
plt.grid(True)
plt.show()
```
